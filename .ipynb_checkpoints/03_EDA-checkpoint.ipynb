## Player Participation and Performance

This section calculates each playerâ€™s total matches, wins, losses, and overall win rate. It also computes win rates by surface type (Hard, Clay, Grass) to identify players who perform better on specific surfaces.

```python
# Count total matches played by each player (as winner or loser)
player_match_counts = (
    df['winner_name'].value_counts() + df['loser_name'].value_counts()
).sort_values(ascending=False)

# Select top 10 most active players
top_players = player_match_counts.head(10)
top_players
```

```python
# Count wins for each player
wins = df['winner_name'].value_counts()

# Count losses for each player
losses = df['loser_name'].value_counts()

# Combine wins and losses into a single DataFrame
player_stats = (
    pd.DataFrame({'wins': wins, 'losses': losses})
    .fillna(0) # Fill missing values (players with no wins/losses in dataset)
    .astype(int) # Ensure counts are integers
)

# Calculate total matches and win rate per player
player_stats['total'] = player_stats['wins'] + player_stats['losses']
player_stats['win_rate'] = player_stats['wins'] / player_stats['total']

# Sort players by win rate
player_stats = player_stats.sort_values(by='win_rate', ascending=False)
player_stats.head(10)
```

```python
# Count wins by player and surface
surface_win_rates = (
    df.groupby(['winner_name', 'surface']).size()
    .unstack(fill_value=0)
)

# Count total matches by player and surface (wins + losses)
surface_total_matches = (
    df.groupby(['winner_name', 'surface']).size()
    .unstack(fill_value=0) +
    df.groupby(['loser_name', 'surface']).size()
    .unstack(fill_value=0)
)

# Calculate win rate per surface
surface_win_rate = surface_win_rates / surface_total_matches

# Sort players by win rate on Hard courts
surface_win_rate.sort_values(by='Hard', ascending=False)
```

## Correlation Analysis

We explore relationships between numeric features such as aces, double faults, and minutes played.

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Drop columns that are IDs or unnecessary for correlation
drop_cols = ["tourney_id", "winner_id", "loser_id", "Unnamed: 0"]
numeric_df = numeric_df.drop(columns=[c for c in drop_cols if c in numeric_df.columns])
```

```python
# Compute correlation matrix
corr_matrix = numeric_df.corr()
```

```python
# Import seaborn for plotting
import seaborn as sns

# Create figure with larger size for readability
plt.figure(figsize=(14, 10))

# Plot heatmap of correlations
# - cmap: color scheme
# - center=0: zero correlation is centered in color map
# - annot=False: do not annotate with numbers (can be turned on if desired)
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, annot=False, fmt=".2f")

# Add title and display
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
